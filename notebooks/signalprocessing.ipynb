```python
import matplotlib.pyplot as plt
from PIL import Image
import skimage
import numpy as np
import torch
from torchvision.transforms import Resize, Compose, ToTensor, Normalize
```

```python
sidelength = 512
```

```python
img = Image.fromarray(skimage.data.camera())
transform = Compose([
        Resize(sidelength),
        ToTensor(),
        Normalize(torch.Tensor([0.5]), torch.Tensor([0.5]))
    ])
img = transform(img).view(sidelength, sidelength)
```

```python
# spectrum = torch.fft.ifftshift(img)
spectrum = torch.fft.fft2(img)
spectrum = torch.fft.fftshift(spectrum)

log_spectrum = torch.log(spectrum.abs())

plt.imshow(log_spectrum, cmap='gray');
```

```python
spectrum = np.fft.fftshift(np.fft.fft2(img))
plt.imshow(np.log(abs(spectrum)), cmap='gray');
```
